### Get Pre-Requisite Topics

```python
def get_prerequisite_topics(topic,k=5):
    prompt = f""" A student wants to learn about {topic}.
    List the {k} most important prerequisite topics needed to understand it in learning order.
    Only return the numbered list of topics without any additional text.
    """

    response = client.generate_content(prompt)
    

    topics = response.text.split("\n")
    final_topics = []

    for topic in topics:
        topic = topic.strip()
        if len(topic) > 0:
            if "." in topic:
                topic = topic.split(".")[1].strip()
        
            final_topics.append(topic)

    
    return final_topics[:k]
```

### Generate Article

```python
def generate_article(topic):
    prompt = f""" Write a detailed article about {topic} that is suitable for a beginner.
    Avoid using heavy jargons.
    Maximum 250 words.
    """

    response = client.generate_content(prompt)
    return response.text.strip()
```

### Input Topic


```python
topic = input("Enter a topic you want to learn about: ")
prerequisites = get_prerequisite_topics(topic)

print(f"Prerequisite topics to learn {topic}:")
for idx, prereq in enumerate(prerequisites):
    print(f"{idx+1}. {prereq}")

print("\nGenerating articles for each prerequisite topic...\n")
for prereq in prerequisites:
    article = generate_article(prereq)

    print(f"{prereq}:\n{article}\n {'-'*50}\n\n")

    file_name = prereq.lower().replace(" ","_") + ".txt"
    with open(file_name, "w") as f:
        f.write(article)

    
```
