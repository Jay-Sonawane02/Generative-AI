```python
import numpy as np
import pandas as pd
from google import genai
from tqdm.auto import tqdm
from sentence_transformers import SentenceTransformer

client = genai.Client(api_key="")

model = SentenceTransformer("Qwen/Qwen3-Embedding-0.6B")
df = pd.read_csv('/content/datasets/GFG Articles/data/0.csv')
```

```python
df = df[:200]

embeddings = []
titles = []
content = []

for title, article in tqdm(zip(df['title'],df['article'])):

  embedding = model.encode(article)

  titles.append(title)
  embeddings.append(embedding)
  content.append(article)

data =pd.DataFrame()

data['title'] = titles
data['content'] = content
data['embeddings'] = embeddings
```

```python
qn = 'adobe interview qns?'

def generate_response(qn, k = 5):

  embedding = model.encode(qn)
  article_embeddings = np.array(list(data['embeddings']))
  similarity_scores = np.dot(article_embeddings, embedding) / (np.linalg.norm(article_embeddings, axis=1) * np.linalg.norm(embedding))
  data['similarity'] = similarity_scores
  similarity_search_results = data.sort_values(by='similarity', ascending=False)
  context = '\n\n'.join(similarity_search_results['content'].head(k).values)

  prompt = '''You are a helpfull chatbot who can answer question based on just the given context. YOu can not answer as per your own knowldge
  you context is this : ''' + context + ''' and your question is ''' + qn

  response = client.models.generate_content( model = "gemini-2.5-flash", contents = prompt)

  return response.text
```

```python
print(generate_response('what is ai', k = 3))
```
